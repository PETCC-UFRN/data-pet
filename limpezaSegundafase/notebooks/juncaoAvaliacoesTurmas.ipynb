```python
import pandas as pd
```

```python
dfTurmas = pd.read_csv("../CSVs/Finais/turmas_geral.csv", sep=";")
dfAvaliacoesDocentes = pd.read_csv("../CSVs/Finais/docentes_e_avaliação_criptografado.csv", sep=";")
```

```python
dfTurmas = dfTurmas.drop(labels="ano", axis=1)
dfTurmas = dfTurmas.drop(labels="periodo", axis=1)
dfTurmas = dfTurmas.drop(labels="total_alunos", axis=1)

dfTurmas.info()
```

```python
df_merged = pd.merge(
    left=dfAvaliacoesDocentes,          # DataFrame da esquerda (base)
    right=dfTurmas,       # DataFrame da direita (a ser unido)
    on='id_turma',          # Coluna de junção
    how='left'              # Tipo de JOIN (LEFT, RIGHT, INNER, OUTER)
)
```

```python
df_merged.info()
```

```python
df_nulos = df_merged[df_merged['media_final_geral'].isna()]
```

```python
df_merged = df_merged[
    ~((df_merged['ano'] == 2023) | (df_merged['ano'] == 2024))
]

df_merged.sort_values('ano', ascending=False)
```

```python
df_merged.to_csv(
    '../CSVs/Finais/avaliacoes_Docentes_Turmas.csv',  # Nome do arquivo de saída
    index=False,                # Não incluir o índice no arquivo
    sep=';',                    # Separador (padrão é vírgula)
    header=True,                # Incluir cabeçalho
)
```
